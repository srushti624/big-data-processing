```python
f = open("watches.html", "r")
webpage = f.read()
f.close()
```

```python
soup=BeautifulSoup(webpage,"html.parser")
soup
```

```python
soup.prettify()
```

```python
links=soup.find_all("span",attrs={'class':"a-size-base-plus a-color-base a-text-normal"})
```

```python
links
```

```python
for i in links:
  print(i.text)
```

```python
price=[]
for i in soup.find_all("span",attrs={'class':"a-price-whole"}):
 price.append(i.text.strip())
price=np.transpose(price)
print(price)
```

```python
title=[]
for i in soup.find_all("span",attrs={'class':"a-size-base-plus a-color-base a-text-normal"}):
  title.append(i.text)
title=np.transpose(title)
print(title)
```

```python
ratings=[]
for i in soup.find_all("i",attrs={'class':"a-icon a-icon-star-small a-star-small-4 aok-align-bottom"}):
  ratings.append(i.text.strip())
ratings=np.transpose(ratings)
print(ratings)
```

```python
bought = []
for i in soup.find_all('span', attrs={'class': "a-size-base a-color-secondary"}):
    if i.text.strip() == 'M.R.P:':
        bought.append('Not Available')
    else:
        bought.append(i.text.strip())

bought=np.transpose(bought)
print(bought)
```

```python
company=[]
for i in soup.find_all('span',attrs={'class':"a-size-base-plus a-color-base"}):
  company.append(i.text)
company=np.transpose(company)
print(company)
```

```python
min_length=min(len(company),len(title),len(ratings),len(price),len(bought))
company=company[:min_length]
title=title[:min_length]
ratings=ratings[:min_length]
price=price[:min_length]
bought=bought[:min_length]
```

```python
data_dict={'COMPANY':company,'TITLE':title,'RATINGS':ratings,'PRICE':price,'BOUGHT':bought}
```

```python
df=pd.DataFrame(data_dict)
```
